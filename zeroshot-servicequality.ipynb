```python
import numpy as np 
import pandas as pd
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",model="facebook/bart-large-mnli")
```

# ***2-Star Hotel***

```python
star_2 = pd.read_csv("/kaggle/input/hotel-review/2_star_hotel_clear.csv")
star_2_sequence = star_2.iloc[:,2]
```

```python
iteration = 0
responseList = []  
temp_response = []
res = []
for response in star_2_sequence:
    response_labels = ['Tangible', 'Reliability', 'Responsiveness', 'Empathy', 'Assurance']
    result = classifier(response, response_labels)
    for key, val in result.items():
        responseList.append([key, val])
    iteration += 1
    print(str(iteration) + "/" + str(len(star_2_sequence)))
```

```python
for i in range(0,len(responseList)):
    if i%3 == 1:
        temp_response.append(responseList[i])

for j in range(0,len(temp_response)):
    res.append(temp_response[j][1][0])
```

```python
star_2['result'] = res
```

```python
star_2.head()
```

```python
star_2_positive = star_2.loc[star_2['rate'] >= 4]
star_2_positive
```

```python
star_2_positive.groupby(['result']).count()
```

```python
star_2_neutral = star_2.loc[star_2['rate'] == 3]
star_2_neutral
```

```python
star_2_neutral.groupby(['result']).count()
```

```python
star_2_negative = star_2.loc[star_2['rate'] <= 2]
star_2_negative
```

```python
star_2_negative.groupby(['result']).count()
```
